```python
#1. Importing Required Libraries
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
#2. Download Required NLTK Resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
#3. Input Text
text = "Natural Language Processing is an amazing field of Artificial Intelligence that enables computers to understand human language."
```

```python
print(text)
```

```python
#4. Step 1 – Tokenization
words = word_tokenize(text)
print("Tokenized Words:", words)
```

```python
#5. Step 2 – Stop Word Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print("After Stop Word Removal:", filtered_words)
```

```python
#6. Step 3 – Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("After Stemming:", stemmed_words)
```
